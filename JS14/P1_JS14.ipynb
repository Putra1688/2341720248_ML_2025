# **PRAKTIKUM 1**

## 1) Import Library

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## 2) Preprocessing Data

### 2.1) Preprocessing Data Training

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/Colab Notebooks/Machine Learning/JS14-CNN/dataset/dataset/training_set',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

### 2.2) Preprocessing Data Testing

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/Colab Notebooks/Machine Learning/JS14-CNN/dataset/dataset/test_set',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

## 3) Pembuatan MOdel CNN

### 3.1) Inisiasi Model CNN

```python
cnn = tf.keras.models.Sequential()
```

### 3.2) Pembuatan Layer Konvolusi

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

### 3.3)Pembuatan Layer Pooling 1

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

### 3.4) Pembuatan Layer Konvolusi 2 dan Pooling 2

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

### 3.5) Flattening

```python
cnn.add(tf.keras.layers.Flatten())
```

### 3.6) Fully Connected Layer 1 (input)

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

### 3.7) Fully Connected Layer 2 (Outpu)

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

### 3.8) Compile Model CNN

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

## 4) Fit CNN

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```
